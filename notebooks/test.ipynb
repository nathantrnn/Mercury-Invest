```python
import os
import pandas as pd

# Specify the paths to both bronze layer directories
bronze_layer_dirs = ['../data-lake/silver/macro', '../data-lake/silver/stocks']

# Iterate through each directory
for bronze_layer_dir in bronze_layer_dirs:
    print(f"Processing directory: {bronze_layer_dir}\n")

    # Ensure the directory exists
    if not os.path.exists(bronze_layer_dir):
        print(f"Directory {bronze_layer_dir} does not exist. Skipping...\n")
        continue

    # Iterate through all files in the directory
    for file_name in os.listdir(bronze_layer_dir):
        if file_name.endswith('.csv'):
            try:
                file_path = os.path.join(bronze_layer_dir, file_name)

                # Read the CSV file
                data = pd.read_csv(file_path)

                # Display file name, column names, data types, and sample data
                print(f"File: {file_name}")
                print("Column Names and Data Types:")
                print(data.dtypes)
                print("\nSample Data:")
                print(data.head(3))
                print("\n" + "=" * 50 + "\n")

            except Exception as e:
                # Handle errors that occur during file reading
                print(f"Error reading file {file_name}: {e}")

    print("=" * 80 + "\n")
```
